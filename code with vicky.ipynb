```python
import requests
from bs4 import BeautifulSoup


# Making a GET request
r = requests.get('https://www.airtel.in/myplan-infinity/')
print(r.text)
# Parsing the HTML
soup = BeautifulSoup(r.content)
#print(soup.prettify())
#title = soup.title
#print(title)
```

```python
title = soup.title
print(title)
```

```python
print(type(title))  # 1. tag
print(type(title.string)) # NavigableSring
print(type(soup))  # Beautifulsoup
```

```python
# get all the paragraphs from the page

paras = soup.find_all('p')
print(paras)
```

```python
# get all the anchor tags from the page

anchor = soup.find_all('div')
print(anchor)
```

```python
# get first element in the HTML page
print(soup.find('p'))
```

```python
# get classes of any element in the HTML Page
print(soup.find('p')['class'])
```

```python
# find all the elements with class text-sm
print(soup.find_all('p', class_='text-sm'))
```

```python
# get the text from the tags/soup


#print(soup.find('p').get_text())
print(soup.get_text())
```

```python
#  get all the anchor tags from the page
anchors = soup.find_all('a')


# get all the links on the page:
for link in anchors:
    print(link.get('href'))
```

```python
#  get all the anchor tags from the page
anchors = soup.find_all('a')
all_links = set()

# get all the links on the page:
for link in anchors:
    if link.get('href') != '#':
        linkText = "https://www.github.com/codewithharry" + link.get('href')
        all_links.add(link)
        print(linkText)
```

```python
import requests
from bs4 import BeautifulSoup


headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }
```

```python
url = "https://www.codewithharry.com/"
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
divariable = soup.find("div", {"class": "flex bg-white h-96 container mx-auto"})
print(divariable)
```

```python
x2 = divariable.find('h2',{"class":"text-3xl font-semibold text-gray-800 md:text-4xl"})
```
